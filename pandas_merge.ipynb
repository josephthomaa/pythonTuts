```python
import pandas as pd
import numpy as np
import glob
import os
import dtale
import matplotlib
```

```python
files = glob.glob("archive/stocks/*.CSV")
df = pd.DataFrame()
for file in files:
    f_name, _ = os.path.splitext(os.path.basename(file))
    year = f"20{f_name[-2:]}"
    csv = pd.read_csv(file)
    csv["year"] = year
    df = df.append(csv)
```

```python
df.head()
# d = dtale.show(df)
# d.open_browser()
```

```python
stock_change = pd.DataFrame()
for name, group in df.groupby("SC_CODE"):
    try:
        print(name)
        group.sort_values('year', ascending=True, inplace=True)
        n = group['CLOSE'].count()-1
        val_change = group.iloc[n].CLOSE - group.iloc[0].CLOSE
        val_change_perc = round((val_change/group.iloc[0].CLOSE) * 100, 2)
        group["perc_change"] = val_change_perc
        stock_change = stock_change.append(group.iloc[n])
    except Exception as e:
        print("here",e)
        print(name)
        print(group['CLOSE'].count())
print("done")
```
